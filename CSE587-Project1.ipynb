```python
df_normalized = (df - df.min())/ (df.max() - df.min())
```

```python
df_normalized.head(5)
```

```python
# K_means Custering
# First we convert the dataframe to a corresponding numpy array for custering purposes
X = df_normalized.to_numpy()
```

```python
X.shape
```

```python

k = 12
cluster_centers = np.random.random((k, X.shape[1]))
cluster_centers.shape
```

```python
from scipy.spatial import distance
def create_empty_clusters(k):
    clusters = []
    for i in range(k):
        clusters.append([])

    return clusters

def assigning_clusters(X,cluster_centers):
    clusters  = create_empty_clusters(cluster_centers.shape[0])
    for i,rowx in enumerate(X):
        dist = []
        for j,rowc in enumerate(cluster_centers):
            dist.append(distance.euclidean(rowx, rowc))
        dist = np.asarray(dist)
        min_dist_index = np.argmin(dist)
        clusters[min_dist_index].append(rowx)
    clusters = np.asarray(clusters)
    return clusters
    
```

```python
for epochs in range(1,100):
    clusters = assigning_clusters(X,cluster_centers)
    print(epochs)
    for i, row in enumerate(clusters):
        sum_elements = np.zeros(X.shape[1])
        for j in row:
            sum_elements = np.add(j, sum_elements)
        cluster_centers[i] = sum_elements * (1/sum_elements.shape[0])
#print(clusters)   
    
```
